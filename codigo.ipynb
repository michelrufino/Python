Operações basicas

```python
#Soma
print(3+3)
#Subtração 
print(2-6)
#Multiplicação
print(2*8)
#Divisão
print(5/8)
#mode
print(8%6)
#elevação
print(7**8)
```

Operações com texto String

```python
# quando coloca o sinal de + está concatenando uma string com a outra 
print("My name is " + "Michel Rufino")

# Palavra in verifica se um texto está está contido dentro de outro texto
# retorna com true ou false
print("Rufino" in "Rufino") # trabalha com fator case sensitive

```
